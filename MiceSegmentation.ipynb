```python
import deeplabcut
import pickle5 as pickle
```

```python
ProjectFolderName = 'miceBodyPartSegmentation-idaapayo-2024-04-29'
VideoType = 'mp4' #, mp4, MOV, or avi, whatever you uploaded!

videofile_path = ['/content/drive/My Drive/'+ProjectFolderName+'/videos/']

path_config_file = '/content/drive/My Drive/'+ProjectFolderName+'/config.yaml'
path_config_file
```

```python

shuffle = 1

deeplabcut.create_multianimaltraining_dataset(path_config_file, Shuffles=[shuffle], net_type="resnet_101",windows2linux=False)
```

```python
deeplabcut.train_network(path_config_file, shuffle=shuffle, displayiters=100,saveiters=1000, maxiters=20000, allow_growth=True)
```

```python

#let's evaluate first:
deeplabcut.evaluate_network(path_config_file,Shuffles=[shuffle], plotting=True)
#plot a few scoremaps:
deeplabcut.extract_save_all_maps(path_config_file, shuffle=shuffle, Indices=[0])
```

```python
#Video to analyze
novel_video_filePath = '/content/drive/MyDrive/miceBodyPartSegmentation-idaapayo-2024-04-29/videos/novelTestVideo.mp4'
deeplabcut.analyze_videos(path_config_file,novel_video_filePath, shuffle=shuffle, videotype=VideoType)
```

```python
deeplabcut.create_video_with_all_detections(path_config_file, [novel_video_filePath], shuffle=shuffle)
```

```python
numAnimals = 4 #how many animals do you expect to find?
tracktype= 'ellipse'
deeplabcut.convert_detections2tracklets(path_config_file, novel_video_filePath, videotype=VideoType,
                                        shuffle=shuffle, overwrite=True)

deeplabcut.stitch_tracklets(path_config_file, novel_video_filePath, shuffle=shuffle, track_method=tracktype, n_tracks=numAnimals)
```

```python
deeplabcut.filterpredictions(path_config_file,
                                 novel_video_filePath,
                                 shuffle=shuffle,
                                 videotype=VideoType,
                                 track_method = tracktype)
```

```python

deeplabcut.plot_trajectories(path_config_file, novel_video_filePath, videotype=VideoType, shuffle=shuffle, track_method=tracktype)
```

```python
deeplabcut.create_labeled_video(path_config_file,
                                novel_video_filePath,
                                shuffle=shuffle,
                                color_by="bodypart",
                                videotype=VideoType,
                                save_frames=False,
                                filtered=True)
```
